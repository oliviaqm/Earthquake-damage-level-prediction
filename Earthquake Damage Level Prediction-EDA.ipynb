```python
# Segregate numerical variables into groups for easy analysis
building_spec = ['count_floors_pre_eq', 'age', 'area_percentage', 'height_percentage']
building_occupancy_density = ['count_families']
```

```python
# Define a custom function for easy and efficient analysis of numerical univariate 

def UVA_numeric(data, var_group):
  '''
  Univariate_Analysis_numeric
  takes a group of variables (INTEGER and FLOAT) and plot/print all the descriptives and properties along with KDE.

  Runs a loop: calculate all the descriptives of i(th) variable and plots/prints it
  '''

  size = len(var_group)
  #looping for each variable
  for index,var in enumerate(var_group):
    
    # calculating descriptives of variable
    min = data[var].min()
    max = data[var].max()
    range = data[var].max()-data[var].min()
    mean = data[var].mean()
    median = data[var].median()
    st_dev = data[var].std()
    skew = data[var].skew()
    kurt = data[var].kurtosis() # kurtosis>3: higher chance of outliers than normal dist / kurtosis<3: lower chance of outliers than normal dist

    # calculating points of standard deviation
    points = mean-st_dev, mean+st_dev

    #Plotting the variable with every information
    plt.figure(figsize = (7*size,3), dpi = 100)
    
    plt.subplot(1,size,index+1) # rows, cols, index
    sns.kdeplot(data[var], shade=True)
    sns.lineplot(points, [0,0], color = 'green', label = "std_dev")
    sns.scatterplot([min,max], [0,0], color = 'orange', label = "min/max")
    sns.scatterplot([mean], [0], color = 'red', label = "mean")
    sns.scatterplot([median], [0], color = 'blue', label = "median")
    plt.xlabel('{}'.format(var), fontsize = 20)
    plt.ylabel('density')
    plt.title('std_dev = {}; kurtosis = {};\nskew = {}; range = {}\nmean = {}; median = {}'.format((round(points[0],2),round(points[1],2)),
                                                                                                   round(kurt,2),
                                                                                                   round(skew,2),
                                                                                                   (round(min,2),round(max,2),round(range,2)),
                                                                                                   round(mean,2),
                                                                                                   round(median,2)))
    
    plt.show();
```

```python
UVA_numeric(data, building_spec);
```

```python
UVA_numeric(data, building_occupancy_density);
```
